```python
# 1 Read the data from the CSV file into a DataFrame and display the data to get a count of the number of rows.
data = pd.read_csv("ramen-ratings.csv")

#Convert Stars to numeric
data['Stars'] = data['Stars'].replace('Unrated', '')
data['Stars'] = pd.to_numeric(data['Stars'])

data
```

```python
# 2 Add a column to the DataFrame that stores the average rating for each country. 
# Then, display the first five rows and verify that the column was added and that its values appear to be correct.
data['Average Rating'] = data.groupby('Country')['Stars'].transform('mean')
data.head()
```

```python
# 3 Group the data by the Country and Style columns and take the mean. 
# Then, unstack the Style column and display the data for the Stars column.
group = data.groupby(['Country', 'Style']).mean().reset_index()
group = group.set_index(['Country', 'Style'])['Stars'].unstack()
group
```

```python
# 4 Write a function that takes a row as input and returns a shortened version of the values in the Variety column. 
# The values should consist of the first two words of the variety name, followed by a space and an ellipsis
def short_variety(row):
    words = row['Variety'].split()
    return ' '.join(words[:2]) + ' ...'
```

```python
# 5 Apply the function you just wrote to the DataFrame, and store the results in a column named ShortVariety. 
# Then, display the first five rows to be sure this worked
data['ShortVariety'] = data.apply(short_variety, axis=1)
data.head()
```

```python
# 6 Set an index on the Brand column and then display the first five rows of data.
data = data.set_index('Brand')
data.head()
```
