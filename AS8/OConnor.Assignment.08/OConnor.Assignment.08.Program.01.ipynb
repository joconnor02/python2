```python
# 1 Read the data from the CSV file into a DataFrame and display the data to get a count of the number of rows.
data = pd.read_csv("ramen-ratings.csv")
data
```

```python
# 2 Rename the Stars column to Rating.
data = data.rename(columns={'Stars': 'Rating'})
```

```python
# 3 Convert the Style column to the category type.
data['Style'] = data['Style'].astype(dtype='category')
data.dtypes
```

```python
# 4 Drop the country column
data = data.drop('Country', axis=1)
```

```python
# 5 Check that the changes were applied to the DataFrame
data.head()
```

```python
# 6 Rename the Brand column to Company and the Variety column to Product.
data = data.rename(columns={'Brand': 'Company'})
data = data.rename(columns={'Variety': 'Product'})
```

```python
# 7 Use the duplicated() method to display all rows with duplicate Company/Product combinations
data[data.duplicated(['Company', 'Product'], keep=False)]
```

```python
# 8 Drop all rows with duplicate Company/Product combinations except for the first row in each set. 
# Then, display the data to get a count of the number of rows, and compare that with the original number of rows to see how many were dropped.
data = data.drop_duplicates(['Company', 'Product'], keep='first')

#2580 rows original - 2520 rows after drop = 60 rows dropped
```

```python
# 9 Drop any rows that have missing values.
data = data.dropna()
```
