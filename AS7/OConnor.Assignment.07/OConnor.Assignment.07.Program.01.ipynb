## Tasks

```python
# 1 Read the data from the CSV file into a DataFrame and display the first five rows.
data = pd.read_csv("ramen-ratings.csv")

#Convert Stars to numeric
data['Stars'] = data['Stars'].replace('Unrated', '')
data['Stars'] = pd.to_numeric(data['Stars'])

data.head()
```

```python
# 2 Create a bar plot to show the Stars by Style with a Seaborn specific method.
sns.barplot(y='Stars', x='Style', data=data)
```

```python
# 3 Create a bar plot to show the Stars by Style with a Seaborn generic method.
theplot = sns.catplot(kind='bar', y='Stars', x='Style', data=data)
```

```python
# 4 Modify the plot you just created to use a custom title and x-axis label.
theplot = sns.catplot(kind='bar', y='Stars', x='Style', data=data).set(xlabel='Ramen Style', title='Average Rated Stars by Ramen Style')
```

```python
# 5 Use the query() method to filter the data to contain only data for Japan, India, Taiwan, and the U.S. 
# Plot the stars by style for each country on a subplot with two subplots per row.
toplot = data.query("Country==['Japan', 'Taiwan', 'India', 'USA']")
theplot = sns.barplot(data=toplot, x='Country', y='Stars')
```
