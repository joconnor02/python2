## Packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
```

## Tasks

```python
# 1 Read the data from the CSV file into a DataFrame and display the first five rows.
data = pd.read_csv("ramen-ratings.csv")
data.head()

#Convert the Stars column values from string objects to float objects
data['Stars'] = data['Stars'].replace('Unrated', '')
data['Stars'] = pd.to_numeric(data['Stars'])
```

```python
# 2 Create a histogram for the Stars column using the number of bins necessary to display the ratings for each quarter of a star
plot_data = data['Stars']
plot_data.plot.hist(bins=20)
```

```python
# 3 Create a KDE plot for the Stars column.
plot_data = data['Stars']
sns.kdeplot(plot_data)
```

```python
# 4 Create a bar plot that shows the average number of stars by ramen style. Hint: Youll need to group the data before you plot it.
plot_data = data[['Style', 'Stars']]
plot_data.groupby('Style')['Stars'].mean().plot.bar()
```

```python
# 5 Create a bar plot that shows the total number of reviews for each ramen type.
plot_data = data['Style'].value_counts()
plot_data.plot.bar()
```

```python
# 6 Create a bar plot that shows the highest score given to each ramen type.
plot_data = data[['Style', 'Stars']].groupby('Style').max()
plot_data.plot.bar()
```
